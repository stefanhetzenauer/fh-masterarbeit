```python
import pickle
import os
import pandas as pd

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report
from sklearn.model_selection import train_test_split
```

```python
def split_data(path, split=0.3):
    df = pd.read_csv(path, sep=';', encoding='utf-8')
    y = df['label']
    X = df.drop(columns=['label'])
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=split)

    return X_train, X_test, y_train, y_test
```

```python
def extract_hyperparameters(MAIN_PATH):
    df = pd.DataFrame()
    for path1 in os.listdir(MAIN_PATH):
        for path2 in os.listdir(MAIN_PATH + path1):
            for path3 in os.listdir(MAIN_PATH + path1 + '/' + path2):
                #print(MAIN_PATH, path1, path2, path3)
                for cms in os.listdir(MAIN_PATH + path1 + '/' + path2 + '/' + path3):
                    if '_model.pkl' in cms and 'binary' not in cms:
                        p = MAIN_PATH + path1 + '/' + path2 + '/' + path3 + '/' + cms
                        with open(p, 'rb') as data:
                            model = pickle.load(data)
                                                        
                            #df = df.append({
                            #    'path': p,
                            #    'params': model.best_params_,
                            #    'key': f'{path1}-{path2}-{path3}',
                            #}, ignore_index=True)
                    if '_cm.csv' in cms and 'norm' not in cms and 'binary' not in cms:
                        p = MAIN_PATH + path1 + '/' + path2 + '/' + path3 + '/' + cms
                        cm = pd.read_csv(p, sep=';', encoding='utf-8')
                        tp = cm['0'][0] + cm['1'][1] + cm['2'][2]
                        a1 = cm['0'][0] + cm['0'][1] + cm['0'][2]
                        a2 = cm['1'][0] + cm['1'][1] + cm['1'][2]
                        a3 = cm['2'][0] + cm['2'][1] + cm['2'][2]
                        
                        acc = tp / (a1 + a2 + a3)
                        print(p, acc, f'{path1}-{path2}-{path3}')
                        
                        df = df.append({
                                'path': p,
                                'acc': acc,
                                'key': f'{path1}-{path2}-{path3}',
                        }, ignore_index=True)
    return df
```

```python
soa = extract_hyperparameters('SOA/')
soa_ = extract_hyperparameters('SOA+/')
all_ = extract_hyperparameters('ALL/')
```

```python
params = pd.merge(pd.merge(soa, soa_, on='key', how='left'), all_, on='key', how='left')
```

```python
params = params.rename(columns={'acc_x': 'SOA', 'acc_y': 'SOA+', 'acc': 'ALL'})
```

```python
del params['path_x']
del params['path_y']
del params['path']
```
